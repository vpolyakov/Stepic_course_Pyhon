```python
with open('dataset_3363_4.txt') as text:
    lines =[line.strip().split(';') for line in text.readlines()]
lines
```

```python
math1 = 0
phisic2 = 0
russian3 = 0
it = 0
for line in lines:
    print((int(line[1]) + int(line[2]) + int(line[3])) / 3)
    math1 += int(line[1])
    phisic2 += int(line[2])
    russian3 += int(line[3])
    it += 1
print(math1/it, phisic2/it, russian3/it)     
```
